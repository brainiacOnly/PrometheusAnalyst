```python
from time import time
import numpy as np
import pandas as pd
import dateutil
import matplotlib.pyplot as plt
from matplotlib import cm
import datetime

from sklearn import metrics
from sklearn.decomposition import PCA
from scipy.cluster.vq import kmeans
from scipy.spatial.distance import cdist,pdist

from IPython.display import Image
```

```python
sm = pd.read_csv("PrometheusHistory/studentmodule_with_dates.csv",quotechar='"',skipinitialspace=True,escapechar='\\').drop(['Unnamed: 0'],1)
users = pd.read_csv("PrometheusHistory/auth_userprofile.csv",quotechar='"',skipinitialspace=True,escapechar='\\')

sm.created = sm.created.apply(lambda x: dateutil.parser.parse(x).replace(tzinfo=None))
sm.modified = sm.modified.apply(lambda x: dateutil.parser.parse(x).replace(tzinfo=None))
sm.publish_date = sm.publish_date.apply(lambda x: dateutil.parser.parse(x).replace(tzinfo=None))
sm = sm.rename(columns={'student_id':'user_id'})
#users = users.drop(['id','location','language','country','city','courseware','allow_certificate'],1)
users = users.drop(['language','location'],1)
```

```python
cert = pd.read_csv("PrometheusHistory/certificates_generatedcertificate.csv",quotechar='"',skipinitialspace=True,escapechar='\\')
```

# History

### #1 Feature preparation: average time of video watching and test passing

```python
sm['time_diff'] = sm.created - sm.publish_date
sm.head()

from datetime import timedelta
#cut negarive time difference
sm = sm[sm.time_diff >= 0]
sm.time_diff = sm.time_diff.apply(lambda x: x/np.timedelta64(1, 'D'))
sm_videos = sm[sm.module_type == 'video']
sm_problems = sm[sm.module_type=='problem']
```

# !!!Here are undefined std's 

```python
videos = sm_videos.groupby('user_id')['time_diff'].agg([np.mean,np.std])
videos = videos.rename(columns={'mean':'video_mean','std':'video_std'})
problems = sm_problems.groupby('user_id')['time_diff'].agg([np.mean,np.std])
problems = problems.rename(columns={'mean':'problems_mean','std':'problems_std'})
videos
```

```python
def sliceUsers(number):
    dates = sm.publish_date.unique()
    pivot = dates[number]
    res = sm.groupby('user_id')['created'].max()
    res = res[res <= pivot]
    data = sm[sm.user_id.isin(res.index)]
    sm_videos = data[data.module_type == 'video']
    sm_problems = data[data.module_type=='problem']
    videos = sm_videos.groupby('user_id')['time_diff'].agg([np.mean,np.std])
    videos = videos.rename(columns={'mean':'video_mean','std':'video_std'})
    problems = sm_problems.groupby('user_id')['time_diff'].agg([np.mean,np.std])
    problems = problems.rename(columns={'mean':'problems_mean','std':'problems_std'})
    features = videos.join(problems).fillna(0)
    features = cert.join(features,on='user_id')
    
    return features

"""
video0
problem0
video50
problem50
video100
problem100
"""
```

```python
#concatenate videos and problems, kill NaNs
#features = videos.join(problems).fillna(0)
#features = cert.join(features,on='user_id')
features = sliceUsers(1)
plt.scatter(features.video_mean,features.problems_mean,c=features.status.apply(lambda x: 'green' if x =='downloadable' else 'grey'))
plt.show()
```

### #2 Apply K-means algorithm for different k(1-20) and find the percentage of variance explained

```python
##### cluster data into K=1..20 clusters #####
X = features[['video_mean','problems_mean']].values
K_MAX = 20
KK = range(1,K_MAX+1)

KM = [kmeans(X,k) for k in KK]
centroids = [cent for (cent,var) in KM]
D_k = [cdist(X, cent, 'euclidean') for cent in centroids]
cIdx = [np.argmin(D,axis=1) for D in D_k]
dist = [np.min(D,axis=1) for D in D_k]

tot_withinss = [sum(d**2) for d in dist]  # Total within-cluster sum of squares
totss = sum(pdist(X)**2)/X.shape[0]       # The total sum of squares
betweenss = totss - tot_withinss          # The between-cluster sum of squares
```

### #3 Plot percentage of variance explained and division on clusters

```python
##### plots #####
kIdx = 8        # K=9
clr = cm.spectral( np.linspace(0,1,10) ).tolist()
mrk = 'os^p<dvh8>+x.'

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(KK, betweenss/totss*100, 'b*-')
ax.plot(KK[kIdx], betweenss[kIdx]/totss*100, marker='o', markersize=12, 
    markeredgewidth=2, markeredgecolor='r', markerfacecolor='None')
ax.set_ylim((0,100))
plt.grid(True)
plt.xlabel('Number of clusters')
plt.ylabel('Percentage of variance explained (%)')
plt.title('Elbow for KMeans clustering')

ax = fig.add_subplot(122)
for i in range(kIdx+1):
    ind = (cIdx[kIdx]==i)
    ax.scatter(X[ind,0],X[ind,1], s=35, c=clr[i], marker=mrk[i], label='C%d'%i)
plt.legend()
plt.title('K=%d clusters'%KK[kIdx])

plt.show()
```

### #6 Apply DBSCAN algorithm for clustering

```python
from sklearn import cluster
from sklearn.neighbors import kneighbors_graph
from sklearn.preprocessing import StandardScaler

colors = np.array([x for x in 'bgrcmykbgrcmykbgrcmykbgrcmyk'])
colors = np.hstack([colors] * 20)

X = StandardScaler().fit_transform(r_features[['video_mean','problems_mean']])
# estimate bandwidth for mean shift
bandwidth = cluster.estimate_bandwidth(X, quantile=0.3)
# connectivity matrix for structured
connectivity = kneighbors_graph(X, n_neighbors=10)
# make connectivity symmetric
connectivity = 0.5 * (connectivity + connectivity.T)
```

```python
algorithm = cluster.DBSCAN(eps=.08)
#spectral = cluster.SpectralClustering(n_clusters=5, eigen_solver='arpack', affinity="nearest_neighbors")
algorithm.fit(X)
if hasattr(algorithm, 'labels_'):
    y_pred = algorithm.labels_.astype(np.int)
else:
    y_pred = algorithm.predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], color=colors[y_pred].tolist(), s=10)
if hasattr(algorithm, 'cluster_centers_'):
    centers = algorithm.cluster_centers_
    center_colors = colors[:len(centers)]
    plt.scatter(centers[:, 0], centers[:, 1], s=100, c=center_colors)
plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.xticks(())
plt.yticks(())
plt.show()
```
