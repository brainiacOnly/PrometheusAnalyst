```python
ids =[i['category'] for i in data._id.values] 
ids = [i for i in range(0,len(ids)) if ids[i] == 'video']
data.loc[ids]._id.head(1)[10]
```

```python
import dateutil.parser
#len(videodata[videodata.module_id == 'i4x://NAUKMA/101/video/788197a3b7144715a5216ba8271f2eab'])
videosource = data.loc[ids]
videosource['module_id'] = pd.Series(['i4x://' + i[1]._id['org'] + '/' + \
                                      i[1]._id['course'] + '/' + \
                                      i[1]._id['category'] + '/' + \
                                      i[1]._id['name'] \
                                      for i in videosource.iterrows()], \
                                     index=videosource.index) 
video = pd.merge(videodata, videosource, how="left", on='module_id')
video['published'] = pd.Series([ dateutil.parser.parse(i[1].edit_info['published_date']) for i in video.iterrows()],index=video.index)#dateutil.parser.parse(video.edit_info['published_date'])
```

```python
video['time_difference'] = pd.Series([i[1].published.replace(tzinfo=None) - datetime.strptime(i[1].modified,"%Y-%m-%d %H:%M:%S") for i in video.iterrows()],index=video.index)
```

```python
print 'negative difference - ', len(video[video.time_difference < 0])/float(len(video))
print 'positive difference - ', len(video[video.time_difference > 0])/float(len(video))
```
