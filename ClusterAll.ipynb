# History and Finance Management Analysis

```python
from time import time
import numpy as np
import pandas as pd
import dateutil
import matplotlib.pyplot as plt
from matplotlib import cm
import datetime
from datetime import timedelta

from sklearn import metrics
from sklearn.decomposition import PCA
from scipy.cluster.vq import kmeans
from scipy.spatial.distance import cdist,pdist
from sklearn import cluster
from sklearn.neighbors import kneighbors_graph
from sklearn.preprocessing import StandardScaler

from IPython.display import Image
```

### #1 Load both data(history and finance management)

```python
fm_sm = pd.read_csv("PrometheusFinManagement/studentmodule_with_dates.csv",quotechar='"',skipinitialspace=True,escapechar='\\').drop(['Unnamed: 0'],1)
fm_users = pd.read_csv("PrometheusFinManagement/auth_userprofile.csv",quotechar='"',skipinitialspace=True,escapechar='\\')

fm_sm.created = fm_sm.created.apply(lambda x: dateutil.parser.parse(x).replace(tzinfo=None))
fm_sm.modified = fm_sm.modified.apply(lambda x: dateutil.parser.parse(x).replace(tzinfo=None))
fm_sm.publish_date = fm_sm.publish_date.apply(lambda x: dateutil.parser.parse(x).replace(tzinfo=None))
fm_sm = fm_sm.rename(columns={'student_id':'user_id'})
fm_users = fm_users.drop(['id','location','language','country','city','courseware','allow_certificate'],1)
```

```python
h_sm = pd.read_csv("PrometheusHistory/studentmodule_with_dates.csv",quotechar='"',skipinitialspace=True,escapechar='\\').drop(['Unnamed: 0'],1)
h_users = pd.read_csv("PrometheusHistory/auth_userprofile.csv",quotechar='"',skipinitialspace=True,escapechar='\\')

h_sm.created = h_sm.created.apply(lambda x: dateutil.parser.parse(x).replace(tzinfo=None))
h_sm.modified = h_sm.modified.apply(lambda x: dateutil.parser.parse(x).replace(tzinfo=None))
h_sm.publish_date = h_sm.publish_date.apply(lambda x: dateutil.parser.parse(x).replace(tzinfo=None))
h_sm = h_sm.rename(columns={'student_id':'user_id'})
h_users = h_users.drop(['language','location'],1)
```

### #2 Make features for Finance Management

```python
fm_sm['time_diff'] = fm_sm.created - fm_sm.publish_date
fm_sm['time_diff'] = fm_sm.time_diff.apply(lambda x: x/np.timedelta64(1, 'D'))

def makeFeatures(sm):
    if 'time_diff' not in sm.columns:
        sm['time_diff'] = sm.created - sm.publish_date
        sm.time_diff.apply(lambda x: x/np.timedelta64(1, 'D'))
    
    #cut negarive time difference
    sm = sm[sm.time_diff >= 0]
    sm_videos = sm[sm.module_type == 'video']
    sm_problems = sm[sm.module_type=='problem']
    
    videos = sm_videos.groupby('user_id')['time_diff'].agg([np.mean,np.std])
    videos = videos.rename(columns={'mean':'video_mean','std':'video_std'})
    problems = sm_problems.groupby('user_id')['time_diff'].agg([np.mean,np.std])
    problems = problems.rename(columns={'mean':'problems_mean','std':'problems_std'})
    features = videos.join(problems).fillna(0)
    return features

fm_features = makeFeatures(fm_sm)
```

### #3 Make features for History

```python
h_sm['time_diff'] = h_sm.created - h_sm.publish_date
h_sm['time_diff'] = h_sm.time_diff.apply(lambda x: x/np.timedelta64(1, 'D'))

h_features = makeFeatures(h_sm)
```

### #5 Join features

```python
def join_features(fm_features,h_features):
    intersection = fm_features[fm_features.index.isin(h_features.index)]
    print "Number of people, who joined both courses - " + str(len(intersection))
    only_fm = fm_features[~fm_features.index.isin(intersection.index)]
    only_h = h_features[~h_features.index.isin(intersection.index)]
    joined = intersection.apply(lambda x: (fm_features.ix[x.name] + h_features.ix[x.name])/2,1)
    joined = joined.append(only_fm).append(only_h)
    return joined
```

```python
joined = join_features(fm_features[['video_mean','problems_mean']],h_features[['video_mean','problems_mean']])
```

### #6 Apply clustering(k-means) for all users

```python
##### cluster data into K=1..20 clusters #####
X = joined.values
K_MAX = 20
KK = range(1,K_MAX+1)

KM = [kmeans(X,k) for k in KK]
centroids = [cent for (cent,var) in KM]
D_k = [cdist(X, cent, 'euclidean') for cent in centroids]
cIdx = [np.argmin(D,axis=1) for D in D_k]
dist = [np.min(D,axis=1) for D in D_k]

tot_withinss = [sum(d**2) for d in dist]  # Total within-cluster sum of squares
totss = sum(pdist(X)**2)/X.shape[0]       # The total sum of squares
betweenss = totss - tot_withinss          # The between-cluster sum of squares

##### plots #####
kIdx = 5        # K=6
clr = cm.spectral( np.linspace(0,1,10) ).tolist()
mrk = 'os^p<dvh8>+x.'

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(KK, betweenss/totss*100, 'b*-')
ax.plot(KK[kIdx], betweenss[kIdx]/totss*100, marker='o', markersize=12, 
    markeredgewidth=2, markeredgecolor='r', markerfacecolor='None')
ax.set_ylim((0,100))
plt.grid(True)
plt.xlabel('Number of clusters')
plt.ylabel('Percentage of variance explained (%)')
plt.title('Elbow for KMeans clustering')

ax = fig.add_subplot(122)
for i in range(kIdx+1):
    ind = (cIdx[kIdx]==i)
    ax.scatter(X[ind,0],X[ind,1], s=35, c=clr[i], marker=mrk[i], label='C%d'%i)
plt.legend()
plt.title('K=%d clusters'%KK[kIdx])

plt.show()
```

### #7 Apply clustering(k-means) for reduced dataset(users started course during first 2 weeks)

```python
def dateFilter(sm,expr):
    min_entrance = sm.groupby('user_id').apply(lambda x: x.sort('created').head(1))
    
    min_entrance = min_entrance[min_entrance.time_diff.apply(expr)]
    first_users = min_entrance.user_id.tolist()
    reduced_sm = sm[sm.user_id.isin(first_users)]
    
    return reduced_sm
```

### #9 Apply clustering(DBSCAN) for reduced dataset(users started course after first 2 weeks)

```python
colors = np.array([x for x in 'bgrcmykbgrcmykbgrcmykbgrcmyk'])
colors = np.hstack([colors] * 20)

X = StandardScaler().fit_transform(r_joined)
# estimate bandwidth for mean shift
bandwidth = cluster.estimate_bandwidth(X, quantile=0.3)
# connectivity matrix for structured
connectivity = kneighbors_graph(X, n_neighbors=10)
# make connectivity symmetric
connectivity = 0.5 * (connectivity + connectivity.T)
```

```python
algorithm = cluster.DBSCAN(eps=.08)
algorithm.fit(X)
if hasattr(algorithm, 'labels_'):
    y_pred = algorithm.labels_.astype(np.int)
else:
    y_pred = algorithm.predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], color=colors[y_pred].tolist(), s=10)
if hasattr(algorithm, 'cluster_centers_'):
    centers = algorithm.cluster_centers_
    center_colors = colors[:len(centers)]
    plt.scatter(centers[:, 0], centers[:, 1], s=100, c=center_colors)
plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.xticks(())
plt.yticks(())
plt.show()
```
